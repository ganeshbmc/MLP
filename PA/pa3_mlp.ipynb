## Q1

```python
from sklearn.datasets import fetch_california_housing
features, labels = fetch_california_housing(as_frame=True, return_X_y=True)
features.shape
```

## Q2

```python
features.isnull().sum()
```

## Q3

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, random_state=0, shuffle=False, test_size=0.2)
```

```python
# Get first three rows of test set
test_features[:3]
```

## Q4

```python
# Get first sample of training set
train_features[:1]
```

## Q6

```python
test_labels.median()
```

## Q7

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
lin_reg_pipeline = Pipeline([("std_scaler", StandardScaler()),
                             ("lin_reg", LinearRegression())])
lin_reg_pipeline.fit(train_features, train_labels)

# Coefficients or weights 
lin_reg_pipeline[-1].coef_
```

## Q8

```python
# Bias term
lin_reg_pipeline[-1].intercept_
```
