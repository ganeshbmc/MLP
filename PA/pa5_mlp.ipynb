```python
idx_6 = np.where(y_train == '6')
idx_9 = np.where(y_train == '9')

X_6 = X_train[idx_6]
X_9 = X_train[idx_9]

y_6 = y_train[idx_6]
y_9 = y_train[idx_9]

X_6_9_stack = np.concatenate((X_6, X_9), axis=0)
y_6_9_stack = np.concatenate((y_6, y_9), axis=0)
```
