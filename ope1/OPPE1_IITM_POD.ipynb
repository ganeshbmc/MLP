Prepare a pipeline numeric_transformer containing SimpleImputer (strategy="mean") and StandardScaler() (in this sequence). Preprocess the numeric_features of the given data ('Adult_Mortality', 'infant_deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis_B', 'Measles', 'BMI', 'under-five_deaths', 'Polio', 'Total_expenditure', 'Diphtheria', 'HIV AIDS', 'GDP', 'Population', 'Income_composition_of_resources', 'Schooling') using this pipeline. Apply 


```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_features = ['Adult_Mortality', 'infant_deaths', 'Alcohol', 'percentage_expenditure', 
                    'Hepatitis_B', 'Measles', 'BMI', 'under-five_deaths', 'Polio',
                    'Total_expenditure', 'Diphtheria', 'HIV_AIDS', 'GDP', 'Population',
                    'Income_composition_of_resources', 'Schooling']

categorical_features = ['Country', 'Status']


numeric_transformer = Pipeline([('imputer', SimpleImputer(missing_values = np.nan, strategy = 'mean')),
                                ('scaler', StandardScaler())])

categorical_transformer = OneHotEncoder(handle_unknown  = 'ignore')

transformer = ColumnTransformer(transformers = [('num', numeric_transformer, numeric_features),
                                                ('cat', categorical_transformer, categorical_features)],
                                remainder = 'passthrough')
                                

transformer.fit(X_train)

X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)
```

```python
X_train.shape
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

estimator = LinearRegression()
selector = RFE(estimator,
               n_features_to_select = 15
               )

selector.fit(X_train, y_train)
```

```python
selector.ranking_
```

```python
selector.support_
```

```python
from sklearn.linear_model import SGDRegressor
sgdregressor = SGDRegressor(penalty = 'l1', alpha = 0.001, learning_rate = 'constant',
                            eta0 = 0.001, random_state = 42)

sgdregressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = sgdregressor.predict(X_test)
y_predt = sgdregressor.predict(X_train)
mse = mean_squared_error(y_test, y_pred)
mset = mean_squared_error(y_train, y_predt)
ans = np.abs(mse - mset)
ans
```
