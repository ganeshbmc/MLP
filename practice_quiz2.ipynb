```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline

mms = MinMaxScaler()
knnr = KNeighborsRegressor(n_neighbors=3, metric='euclidean')
```

```python
X = np.array([[1000, 0.5],
                [2000, 0.6],
                [1500, 0.3],
                [990, 1.0],
                [5400, 0.7]])
y = np.array([11, 20, 42, 16, 25])

pipe = Pipeline(steps=[('scaler', mms),
                        ('knnr', knnr)])
pipe.fit(X, y)
```

```python
x_test = np.array([[1700, 0.2]])
pipe.predict(x_test)
```
